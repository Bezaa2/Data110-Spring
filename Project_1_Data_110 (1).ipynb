There are a total of 2102 deaths in this dataset. It is between the years 2007-2021.

```python
plt.figure(figsize=(15, 6))
sns.histplot(df['Leading Cause'], bins=12, kde=True)
plt.title('Leading Cause of Death', fontsize=20)
plt.xlabel('Leading Cause', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.xticks(rotation=50, ha='right')
plt.show()
```

This histogram shows all the different leading causes of death. The graph is a great visual to see what the highest leading causes and lowest leading causes are. The top Leading causes of death in New York are Disease of Heart, malignant neoplasm (cancer), and influenza and pneumonia. The lowest leading cause was tuberculosis, meningitis, pregnancy childbirth and the puerperium and human immunodeficiency virus disease.

```python
plt.figure(figsize=(30, 10))
sns.countplot(x='Year', data=df)
plt.title('Year with the highest Death', fontsize=20)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.show()
```

The counterplot displays the death rate throughout the years 2007-2021. Although there was a steady rate throughout the years with slight increase and decrease, it shows that in 2019 there was a drastic increase in death. 2019 has the highest death rate. 2013 and 2020 have the lowest death rates.

```python
year_2019 = df[df['Year'] == 2019]

plt.figure(figsize=(15, 5))
sns.histplot(year_2019['Leading Cause'], bins=12, kde=True)
plt.title('Leading Cause of Death in 2019', fontsize=20)
plt.xlabel('Leading Cause', fontsize=10)
plt.ylabel('Frequency', fontsize=15)
plt.xticks(rotation=30, ha='right')
plt.show()
```

The highest leading causes of death in 2019 were disease of heart, malignant neoplasm (cancer), diabetes mellitus, and influenza and pneumonia.

```python
plt.figure(figsize=(8,10))
sns.catplot(x='Year', hue='Race Ethnicity', data=df, kind='count', height=9, aspect=1)

plt.title('Death Count by Race Ethnicity Throughout the Years')
plt.xlabel('Year')
plt.ylabel('Count of Deaths')

plt.show()
```

The Catplot shows the death rates throughout the years within the different ethnicities. Looking at the graph, there are equal counts that stay the same throughout the different ethnicities across the years. The only outliers are other races or not stated.

```python
plt.figure(figsize=(8,10))
sns.catplot(x='Year', hue='Sex', data=df, kind='count', height=9, aspect=1)

plt.title('Death Count by Gender Throughout the Years')
plt.xlabel('Year')
plt.ylabel('Count of Deaths')

plt.show()
```

The Catplot illustrates the death rates throughout the years based on genders. Females seem to have more deaths than males. Although there are years which the 2 genders are both equal at certain years like 2010 and 2019.
